```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target  # Add target column

print(df.head())  # See first 5 rows
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
numerical = df.select_dtypes(include=['int64', 'float64']).columns
categorical = df.select_dtypes(include=['object']).columns

print("Numerical Columns:", numerical.tolist())
print("Categorical Columns:", categorical.tolist())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
print("Unique species values:", df['species'].unique())
print("\nValue counts:")
print(df['species'].value_counts())
```

## 🎯 Interview Questions & Answers

### Q1. What is the difference between Structured and Unstructured data?
- **Structured Data:** Organized in rows & columns (Excel, CSV, SQL)
  - Example: Iris dataset with fixed columns
- **Unstructured Data:** No fixed format
  - Example: Images, videos, emails, text messages

### Q2. Explain Numerical vs Categorical data types?
- **Numerical:** Number columns used for calculations
  - Example: sepal length = 5.1, petal width = 0.2
- **Categorical:** Label/group columns
  - Example: species (0=setosa, 1=versicolor, 2=virginica)

### Q3. What are Continuous and Discrete variables?
- **Continuous:** Decimal values in a range
  - Example: sepal length = 5.1, 4.9, 4.7
- **Discrete:** Only whole countable numbers
  - Example: species = 0, 1, 2

### Q4. How do you handle missing values in a dataset?
- Check using: df.isnull().sum()
- Ways to handle:
  - Fill with mean → df.fillna(df.mean())
  - Fill with median → df.fillna(df.median())
  - Drop rows → df.dropna()
- **In Iris Dataset → No missing values found ✅**

### Q5. What does the .describe() function tell you?
- Shows statistical summary of all numerical columns:
  - **count** → total non-null values (150)
  - **mean** → average value
  - **std** → standard deviation (spread of data)
  - **min** → smallest value
  - **max** → largest value
  - **25%, 50%, 75%** → quartile distribution
